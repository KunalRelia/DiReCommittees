```python
import csv
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np
from matplotlib.ticker import FormatStrFormatter
%matplotlib inline
```

```python
file_path = 'HeuristicAlgo/timeout-phi'
df = pd.read_csv(file_path+'.csv', low_memory=False)

#df
```

```python
file_path = 'HeuristicAlgo/runningtime-phi'
df1 = pd.read_csv(file_path+'.csv', low_memory=False)

#df1
```

```python
fig = plt.figure(figsize=(6, 4), dpi=120)
ax = fig.add_subplot(111)


df['Diversity']=df['Diversity'].astype(str)
df1['Diversity']=df1['Diversity'].astype(str)

col='black'
col1='lightcoral'
col1='indianred'
col1='red'
col2='red'


ax2 = ax.twinx()
sns.violinplot(x="Diversity", y="Proportion",data=df1, ax=ax2, linewidth=0.5, fliersize=True, color=col1)
sns.lineplot(x="Diversity", y="Proportion",data=df, ax=ax, linewidth=2,legend=False, color='black')


for i,box in enumerate(ax2.artists):
    box.set_edgecolor(col)
    
    for j in range(i*6,i*6+6):
        line = ax2.lines[j]
        line.set_color(col)
        line.set_mfc(col)
        line.set_mec(col)


sns.set_style(style="whitegrid")


ax.set_xlabel('$\phi$',fontsize=22)
ax.set_ylabel('% timed out',fontsize=22)
ax2.set_ylabel('running time (s)',fontsize=22, color=col1)


#ax.set_xticks([10,50,100,150,200]) 
#ax.set(xticklabels=['10','50','100','150','200'])
ax.set_yticks([0,25,50,75,100]) 
ax2.set_yticks([0,100,200,300,400,500]) 

ax2.set_ylim(-45,575) 
ax.set_ylim(-10,105)

ax2.yaxis.label.set_color(col2)
ax2.tick_params(axis='y', colors=col2)

ax.tick_params(labelsize=15)
ax2.tick_params(labelsize=15)

ax.tick_params(labelsize=22)
ax2.tick_params(labelsize=22)
ax.tick_params(axis='x',labelsize=16)


ax2.grid(False)
ax.grid(False)

plt.tight_layout()
plt.savefig('runningtime-phi-line+box_v7.pdf')
```
